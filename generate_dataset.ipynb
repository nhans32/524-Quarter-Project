**Enron**

```python
enron_formatted = pd.read_csv('data/formattedData/enronFormatted.csv')

enron_formatted_spam = enron_formatted[enron_formatted['label'] == 1]
enron_formatted_ham = enron_formatted[enron_formatted['label'] == 0]

enron_formatted = list(pd.concat([enron_formatted_spam[:numSamples//2], enron_formatted_ham[:numSamples//2]]).itertuples(index=False, name=None))
```
